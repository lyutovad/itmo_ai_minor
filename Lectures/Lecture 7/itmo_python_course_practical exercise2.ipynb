## Переменные Датасета

* Unnamed: 0: Идентификатор строки (возможно, индекс или временное значение).
* Job Id: Уникальный идентификатор вакансии.
* Experience: Требуемый опыт работы (например, "5 to 15 Years").
* Qualifications: Требуемая квалификация (например, "M.Tech", "MBA").
* Salary Range: Диапазон зарплат (например, "$59K-$99K").
* location: Местоположение вакансии.
* Country: Страна, где расположена вакансия.
* latitude: Широта местоположения вакансии.
* longitude: Долгота местоположения вакансии.
* Work Type: Тип занятости (например, "Intern", "Full-time").
* Company Size: Размер компании (количество сотрудников).
* Job Posting Date: Дата публикации вакансии.
* Preference: Дополнительные предпочтения (например, наличие определенных сертификатов или навыков).
* Contact Person: Контактное лицо для связи по вакансии.
* Contact: Контактные данные (телефон или email).
* Job Title: Название должности (например, "Data Scientist").
* Role: Роль, которая требуется на данной позиции (например, "Software Development").
* Job Portal: Источник вакансии (например, LinkedIn, Indeed).
* Job Description: Текстовое описание вакансии (обязанности, задачи, условия работы).
* Benefits: Преимущества и льготы, предлагаемые на данной позиции (например, медстраховка, бонусы).
* skills: Перечень ключевых навыков, необходимых для выполнения работы.
* Responsibilities: Список обязанностей, которые необходимо выполнять на данной позиции.
* Company: Название компании, предоставляющей вакансию.
* Company Profile: Описание компании (сектор, индустрия, профиль).
